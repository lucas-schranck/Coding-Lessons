```python
# CARD
# SUIT,RANK,VALUE-int
import random # for shuffling deck
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs') # saving as Tuple; I don't want to change this
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace') # all global
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11} # dictionary to link ranks with INT and not STR
# I put all this here for readability

game_on = True
```

```python
# create card class first
class Card: # no need for () as we won't use inheritance
    
    def __init__(self,suit,rank): # no need for user to figure the value
        self.suit = suit
        self.rank = rank
        
    def __str__(self): # string method in case human wants to know what's happening
        return self.rank + " of " + self.suit #We may never need to print the contents of the deck during gameplay, but having the ability to see the cards inside it may help troubleshoot any problems that occur during development
```

```python
class Deck:
    
    def __init__(self): # instantiating; no need for all_cards as we don't need user input
        self.deck = [] # starting as empty list
        
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank)) #notice that everytime I create deck, they're in order
    
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n'+ card.__str__() # calls to print str representation of card
        return "The deck has: "+deck_comp
        
    def shuffle(self): # method to shuffle
        random.shuffle(self.deck) # I don't need to return this, because the method is internal in deck
        
    def deal(self): # method to deal cards
        single_card = self.deck.pop() # this grabs the deck attribute, and pop card from that list and then assign to singlecard
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = [] # empty list, same as deck
        self.value = 0 # always start with a value (null or zero or blank)
        self.aces = 0 # tracking aces (they can turn into 1s)
       
    def add_card(self,card):
        # card passed in is from DECK so we can use deal - Deck.deal().
        self.cards.append(card)
        self.value += values[card.rank] # this takes in card obj from deckdeal, append and adjust to the value
        
        # track aces
        if card.rank == 'Ace':
            self.aces += 1 # everytime I have an ace, I track it
        
    def adjust_for_ace(self):
        
        # If you busted and still have ace
        # You take -10 to adjust Ace value
        # You don't actually change Ace value in the dictionary, only in the hand
        while self.value > 21 and self.aces > 0: # Dont need the >0 here, anything other than 0 is True
            self.value -= 10
            self.aces -= 1
            
```

```python
class Chips:
    
    def __init__(self,total=100):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
# Function to take a bet
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
        except:
            print("Sorry, please provide a number (integer)")
        else:
            if chips.bet>chips.total:
                print(f"Sorry, you do not have enought chips! You have: {chips.total}")
            else:
                break
```

```python
# Function for taking hits

def hit(deck,hand):
    
    
    single_card = deck.deal() # grab card
    hand.add_card(single_card) # add card to hand
    hand.adjust_for_ace() # check for ace
```

```python
# Function to prompt player hit or stand

def hit_or_stand(deck,hand):
    global game_on # to check
    
    while True:
        x = input('Hit or Stand? Enter h or s: ')
        
        if x[0].lower() == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print("Player Stands. Dealer's turn.")
            game_on = False
            
        else:
            print("Please, provide 's' or 'h' only.")
            continue
        break
```

```python
# function to display card
# dealer first card is hidden
# first show players and 1 dealer
# second show all

def show_some(player,dealer):
    
    # dealer.cards[0] hidden whereas cards[1] is shown
    
    # Show only one dealer's card
    print("\n DEALER'S HAND: ")
    print("First card hidden.")
    print(dealer.cards[1])
    
    
    # Show all (2 cards) of the player's hand
    print("\n PLAYER'S HAND: ")
    for card in player.cards:
        print(card)
    print(f"\nValue of Player's hand is: {player.value}\n")
    
    
def show_all(player,dealer): # to check who won
    
    # show all the dealers cards
    print("\n DEALER'S HAND: ")
    for card in dealer.cards:
        print(card)
        
    #instead of for loop you could use:
    # *dealer.card,sep='\n'
    # calculate and display the total value (as in J+9 == 19, etc)
    print(f"Value of Dealer's hand is: {dealer.value}")        



    
    
    # show all the players cards
    print("\n Player's hand: ")
    for card in player.cards:
        print(card)
    print(f"Value of Player's hand is: {player.value}")
```

```python
# Function to handle game scenarios

def player_busts(player,dealer,chips):
    print("Player BUSTED!")
    chips.lose_bet()
    

def player_wins(player,dealer,chips):
    print("Player WINS!")
    chips.win_bet()    
    

def dealer_busts(player,dealer,chips):
    print("Dealer BUSTED!")
    chips.win_bet()    

    
def dealer_wins(player,dealer,chips):
    print("Dealer WINS!")
    chips.lose_bet()
    

def push(player,dealer):
    print("Dealer and player tie. Push.")
```

```python
# Logic of the game

while True:
    
    print("Welcome to Blackjack!")
    
    # Setting up the Deck
    deck = Deck()
    deck.shuffle()
    
    # Setting up the hands
    
    player_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    # Setting up the chips
    player_chips = Chips()
    
    # Prompt player for bets
    take_bet(player_chips)
    
    
    # Show cards (but 1 of dealers hidden)
    show_some(player_hand,dealer_hand)
    
    while game_on: # recall var from hit or stand
    # prompt player to hit or stand
        hit_or_stand(deck,player_hand)
    
    # show cards (keep 1 hidden again)
        show_some(player_hand,dealer_hand)
    # Now checking many possibilities, starting with a player bust
    # if player exceeds, player bust
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            
            break
    # If player hasn't busted        
    if player_hand.value <= 21:
    
        while dealer_hand.value < 17: # dealer hits until 17 at least, as in casinos
            hit(deck,dealer_hand)
        
            show_all(player_hand,dealer_hand)
            # dealer busted
            if dealer_hand.value > 21:
                dealer_busts(player_hand,dealer_hand,player_chips)
            # dealer wins    
            elif dealer_hand.value > player_hand.value:
                dealer_wins(player_hand,dealer_hand,player_chips)
            # player wins    
            elif dealer_hand.value < player_hand.value:
                player_wins(player_hand,dealer_hand,player_chips)
            else: # push
                push(player_hand,dealer_hand)
            
    print(f'\n Players total chips are at: {player_chips.total}')
    
    new_game = input("Would you like to play another hand? y/n: -")
    
    if new_game[0].lower() == 'y':
        game_on = True
        continue
    else:
        print('Thank you for playing')
        break
```
