```python
import requests
```

```python
result = requests.get("https://en.wikipedia.org/wiki/Black_hole")
```

```python
type(result)
```

```python
result.text # the whole html is here
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(result.text, "lxml") # bs4 will use lxml as background to parse through html
```

```python
soup
```

```python
# Let's grab the title
```

```python
soup.select('title') # by default it grabs a list because generally you have multiple elements
```

```python
soup.select('title')[0]
```

```python
soup.select('title')[0].getText()
```

```python
site_paragraphs = soup.select('p')
```

```python
site_paragraphs
```

```python
site_paragraphs[0]
```

```python
type(site_paragraphs[0]) # note this is a specific bs4 element
```

```python
#
# NOW GRABBING A CLASS
#
```

```python
soup.select('.mw-body')
```

```python
type(soup.select('.mw-body')[0])
```

```python
first_item = soup.select('.mw-body')[0]
```

```python
first_item.text
```

```python
for item in soup.select('.mw-body'):
    print(item.text)
```
